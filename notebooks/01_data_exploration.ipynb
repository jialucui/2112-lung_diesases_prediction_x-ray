# Exploratory Data Analysis of Chest X-ray Images

```python
import pandas as pdimport matplotlib.pyplot as pltimport seaborn as snsimport os
```

## 1. Data Loading

```python
# Load the dataset
# Assuming images are stored in a directory named 'chest_xray'
image_dir = 'chest_xray/'
image_files = os.listdir(image_dir)
print(f'Loaded {len(image_files)} images.')
```

## 2. Data Visualization

```python
# Sample visualization of an image
# Example: displaying first image
import matplotlib.image as mpimg
img = mpimg.imread(os.path.join(image_dir, image_files[0]))
plt.imshow(img)
plt.axis('off')
plt.show()
```
